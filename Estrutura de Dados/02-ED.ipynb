# TEMPO DE EXECUÇÃO, NOTAÇÃO BIG O, ARRAYS E LISTAS ENCADEADAS

Uma vez que se define um algoritmo, é importante observar seu tempo de execução. 

Nós (que gostamos de economizar tempo e espaço) escolhemos alguns algoritmos em detrimento de outros pela eficiência. 

Muitos fatores influenciam no tempo de execução de um algoritmo:

* O sistema Operacional utilizado
* A linguagem de programação utilizada
* A velocidade do processador
* O modo como o algoritmo foi implementado
* A estrutura de dados escolhida (cenas dos príxios capítulos)

Ter controle sobre todos esses fatores para saber o quão eficiente é um algoritmo é uma tarefa inviável. Por isso, em computação, o tempo de execução de um determinado algoritmo é medida de uma forma diferente que depende basicamente da entrada e da forma como seu algoritmo foi construído. 

## Voltando ao problema das buscas simples e binária...


### Na busca simples: 


Vimos que para uma lista de 100 números, precisaríamos, na pior das hipóteses de 100 tentativas para encontrar o valor procurado.

Se esta lista tivesse 4 bilhões de números, poderíamos precisar de até 4 bilhões de tentativas para encontrar o valor procurado. 🤯🤯🤯🤯

Logo, podemos observar que o número máximo de tentativas é igual ao tamanho $n$ da lista. Isso é chamado de **tempo linear**.

### Na busca binária:

Se a lista tem 100 itens, precisaríamos, na pior das hipóteses de 7 tentativas para encontrar o valor procurado.

Se a lista tem 4 bilhões de números, poderíamos precisar de até 32 tentativas para encontrar o valor procurado. 😱😱😱

Logo, podemos observar que o número máximo de tentativas é igual a 32. Isso é chamado de **tempo logaritmico**.

# Tempo de execução e Notação Big O 

A notação Big O é uma notação especial que diz o quão rápido é um algoritmo. Ela estabelece o tempo de exceução para a pior hipótese. 

Perceba que pra verificarmos que um agoritmo é melhor que outro precisamos verificar a taxa de crescimento desse algoritmo a medida que o tamanho $n$ da entrada aumenta

<img src="../Figures/taxa_crescimento.png" alt= “” width=500 height=200>

Não basta saber quanto tempo um algorito leva para ser executado - você precisa saber se o tempo de execução aumenta conforme a lista aumenta! 🧐

A notação Big O informa o quão rapidamente um algoritmo cresce. Ou seja ela está ligada a taxa de crescimento do algoritmo e não ao tempo necessário pde execução com uma entrada específica.

#### Arrays dinâmicos

No python, temos a possibilidade de trabalhar com **arrays dinâmicos**, que são abstrações sobre arrays, utilizando-se objetos das classes `array` e `list`.

Em **arrays dinâmicos**, o tamanho do array pode ser aumentado ou reduzido conforme necessário. 

Isso é legal, mas.....

```python
# Inicio a lista sem reservar nenhum espaço em memória
l = []

# Adiciono elementos dinamicamente
for i in range(5):
    l.append(i)

l
```

```python
import array as arr

# Inicio a lista sem reservar nenhum espaço em memória
meu_array = arr.array('f', [])

# Adiciono elementos dinamicamente
for i in range(5):
    meu_array.append(i)

meu_array
```

Uma lista encadeada é uma estrutura dedados linear onde cada elemento é armazenado em um nó que armazena também endereços para outros nós da lista. Por isso,cada nó de uma lista pode estar em uma posição diferente da memória, sendo diferente de um array,onde os elementos são armazenados de forma contínua. 

Na forma mais simples, têm-se acesso apenas ao primeiro nó da lista. Em qualquer variação,têm-se acesso a um número constante de nós apenas(o primeiro nó e o último, por exemplo).

Assim,listas não permitem acesso direto a um elemento: para acessar o k-ésimo elemento da lista, deve-se acessar o primeiro,que dá acesso a osegundo, que dá acesso ao terceiro, e assim sucessivamente, até que o (k−1)-ésimo elemento dá acesso ao k-ésimo.

### Listas encadeadas no python

Em Python, há um objeto específico no módulo `collections` que você pode usar para listas encadeadas chamado `deque`.

`collections.deque` usa uma implementação de uma lista encadeada na qual você pode acessar, inserir ou remover elementos do início ou fim de uma lista com desempenho O(1) constante.

```python
from collections import deque
llist= deque(['a','b','c'])
llist
```

Operações rápidas O(1):

```python
llist.append(1)
llist
```

```python
llist.pop()
llist
```

Operações não tão rápidas O(n):

```python
llist.insert(0, "p0")
llist
```

# Ordenação por seleção

Como percebemos, o algoritmo de busca binária que aprendemos só funciona em listas ordenadas. Diante disso, nos vem a pergunta: Como ordenar uma determinada lista de objetos?

Aqui aprenderemos um, dos diversos algoritmos de ordenação que existem: A **ordenação por seleção**.

A maioria das linguagens de programação já tem algoritmos nativos quem implementam a ordenação utilizando métodos mais eficientes que a ordenação por seleção. Porém aprender sobre esse algoritmo irá te ajudar a compreender outro algoritmo de seleção bastante famoso, que veremos mais a diante, o **quicksort**.

Suponhamos que queiramos ordenar uma lista de músicas das mais tocadas as menos tocadas:

Qual seria o tempo de execução desse algoritmo? A resposta é: O($n^2$)

<img src="../Figures/bigOselecao.png" alt= “” width=500 height=300>


### Ordenação por seleção no python:

```python
def ordenacao_selecao(A):
    n = len(A)
    # Percorre o arranjo A.
    for i in range(n):
        # Encontra o elemento mínimo em A.
        minimo = i
        for j in range(i + 1, n):
            if A[minimo] > A[j]:
                minimo = j
        # Coloca o elemento mínimo na posição correta.
        A[i], A[minimo] = A[minimo], A[i]

    return A
```

```python
from random import shuffle
array = [i for i in range(100000)]
shuffle(array)
print(array[:10])

ordenacao_selecao(array)
print(array[:10])
```

A ordenação por seleção é um bom algorítmo com um tempo de execução razoável. Mas existe um algoritmo melhor que ela, chamado quiksort, cujo tempo de execução é O($n \log n$). Conheceremos este algoritmo mais adiante!

```python
# Numpy usa o quiksort por padrão pra ordenar seus arrays!
array =  np.random.randint(100000, size = 100000)
print(array[:10])

ord_array = array.sort()
print(array[:10])
```
