# TEMPO DE EXECUO, NOTAO BIG O, ARRAYS E LISTAS ENCADEADAS

Uma vez que se define um algoritmo,  importante observar seu tempo de execuo. 

Ns (que gostamos de economizar tempo e espao) escolhemos alguns algoritmos em detrimento de outros pela eficincia. 

Muitos fatores influenciam no tempo de execuo de um algoritmo:

* O sistema Operacional utilizado
* A linguagem de programao utilizada
* A velocidade do processador
* O modo como o algoritmo foi implementado
* A estrutura de dados escolhida (cenas dos prxios captulos)

Ter controle sobre todos esses fatores para saber o quo eficiente  um algoritmo  uma tarefa invivel. Por isso, em computao, o tempo de execuo de um determinado algoritmo  medida de uma forma diferente que depende basicamente da entrada e da forma como seu algoritmo foi construdo. 

## Voltando ao problema das buscas simples e binria...


### Na busca simples: 


Vimos que para uma lista de 100 nmeros, precisaramos, na pior das hipteses de 100 tentativas para encontrar o valor procurado.

Se esta lista tivesse 4 bilhes de nmeros, poderamos precisar de at 4 bilhes de tentativas para encontrar o valor procurado. 

Logo, podemos observar que o nmero mximo de tentativas  igual ao tamanho $n$ da lista. Isso  chamado de **tempo linear**.

### Na busca binria:

Se a lista tem 100 itens, precisaramos, na pior das hipteses de 7 tentativas para encontrar o valor procurado.

Se a lista tem 4 bilhes de nmeros, poderamos precisar de at 32 tentativas para encontrar o valor procurado. 

Logo, podemos observar que o nmero mximo de tentativas  igual a 32. Isso  chamado de **tempo logaritmico**.

Uma lista encadeada  uma estrutura dedados linear onde cada elemento  armazenado em um n que armazena tambm endereos para outros ns da lista. Por isso,cada n de uma lista pode estar em uma posio diferente da memria, sendo diferente de um array,onde os elementos so armazenados de forma contnua. 

Na forma mais simples, tm-se acesso apenas ao primeiro n da lista. Em qualquer variao,tm-se acesso a um nmero constante de ns apenas(o primeiro n e o ltimo, por exemplo).

Assim,listas no permitem acesso direto a um elemento: para acessar o k-simo elemento da lista, deve-se acessar o primeiro,que d acesso a osegundo, que d acesso ao terceiro, e assim sucessivamente, at que o (k1)-simo elemento d acesso ao k-simo.

# Ordenao por seleo

Como percebemos, o algoritmo de busca binria que aprendemos s funciona em listas ordenadas. Diante disso, nos vem a pergunta: Como ordenar uma determinada lista de objetos?

Aqui aprenderemos um, dos diversos algoritmos de ordenao que existem: A **ordenao por seleo**.

A maioria das linguagens de programao j tem algoritmos nativos quem implementam a ordenao utilizando mtodos mais eficientes que a ordenao por seleo. Porm aprender sobre esse algoritmo ir te ajudar a compreender outro algoritmo de seleo bastante famoso, que veremos mais a diante, o **quicksort**.

### Ordenao por seleo no python:

```python
def ordenacao_selecao(A):
    n = len(A)
    # Percorre o arranjo A.
    for i in range(n):
        # Encontra o elemento mnimo em A.
        minimo = i
        for j in range(i + 1, n):
            if A[minimo] > A[j]:
                minimo = j
        # Coloca o elemento mnimo na posio correta.
        A[i], A[minimo] = A[minimo], A[i]

    return A
```

```python
from random import shuffle
array = [i for i in range(100000)]
shuffle(array)
print(array[:10])

ordenacao_selecao(array)
print(array[:10])
```

A ordenao por seleo  um bom algortmo com um tempo de execuo razovel. Mas existe um algoritmo melhor que ela, chamado quiksort, cujo tempo de execuo  O($n \log n$). Conheceremos este algoritmo mais adiante!

```python
# Numpy usa o quiksort por padro pra ordenar seus arrays!
array =  np.random.randint(100000, size = 100000)
print(array[:10])

ord_array = array.sort()
print(array[:10])
```
