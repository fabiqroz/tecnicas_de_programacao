# Herana

A herana  uma forma de formar novas classes usando classes que j foram definidas. 
As classes recm-formadas so chamadas de **classes derivadas** ou **sub classe**, as classes das quais derivamos so chamadas de **classes base** ou **super classe**. 

Benefcios importantes da herana so a reutilizao de cdigo e a reduo da complexidade de um programa. 
As classes derivadas (descendentes) substituem ou estendem a funcionalidade das classes base (ancestrais).

11.png

### Vamos ver um exemplo incorporando nosso trabalho anterior na classe *Cachorro*:

```python
class animal:
    def __init__(self):
      print('- Criando um animal.')  
```

### A classe `cachorro` passa a ser uma sub-classe da classe `animal`.

... assim como as classes `ave` e `peixe`

```python
class cachorro(animal):  
  def __init__(self):
    super().__init__()
    print('-- Criando um cachorro')

class ave(animal):  
  def __init__(self):
    super().__init__()
    print('-- Criando uma ave')

class peixe(animal):  
  def __init__(self):
    super().__init__()
    print('-- Criando um peixe')
```

```python
cao = cachorro()
```

```python
p = peixe()
```

Python nos permite verificar se uma classe herda de outra determinada classe por meio da funo <code>issubclass()</code>:

```python
print(issubclass(cachorro,animal))
print(issubclass(ave, animal))
print(issubclass(peixe, animal))
print(issubclass(cachorro, peixe))
```

Agora vamos atribuir os atributos a classe base e classes  descedentes:

```python
class animal:
    def __init__(self, idade, peso):
        self.reino = 'Animal'
        self.filo = 'Vertebrado'
        self.idade = idade
        self.peso = peso
```

```python

class cachorro(animal):  
  def __init__(self, nome, raca, idade, peso, tutor, vacina):
    super().__init__(idade, peso)
    self.classe = 'Mamifero'
    self.especie = 'Canina'
    self.nome = nome
    self.raca = raca
    self.nome_tutor = tutor
    self.status_vacina = vacina

  # Mtodo latir da classe cachorro
  def latir(self):
    print('{} diz AU AU AU!'.format(self.nome))
    
```

```python
cao = cachorro('toby','SDR', 6, 3.76, 'Maria', True)
```

```python
class ave(animal):  
  def __init__(self, idade, peso, especie, habilidade_voo, habilidade_canto):
    super().__init__(idade, peso)
    self.classe = 'Ave'
    self.especie = especie
    self.habilidade_voo = habilidade_voo
    self.habilidade_canto = habilidade_canto
```

```python
ave1 = ave(idade = 1, peso = 0.3, especie='galinha', habilidade_voo=False, habilidade_canto=False)
ave2 = ave(idade = 1, peso = 0.3, especie='pardal', habilidade_voo=True, habilidade_canto=True)
```

```python
class peixe(animal):  
  def __init__(self, idade, peso, especie, tem_escamas):
    super().__init__(idade, peso)
    self.classe = 'Peixe'
    self.especie = especie
    self.tem_escama = tem_escamas
```

```python
p = peixe(idade = 1, peso = 1.5, especie='salmao', tem_escamas=False)
```

```python
# atributos da classe animal
print(p.reino, ave1.reino,cao.reino)
print(p.filo, ave1.filo,cao.filo)
#atrinutos das classes especificas
print(p.classe, ave1.classe,cao.classe)
print(p.especie, ave1.especie,cao.especie)
# atributos da classe animal
print(p.peso, ave1.peso,cao.peso)
print(p.idade, ave1.idade,cao.idade)
#atrinutos das classes especificas
print(p.tem_escama, ave1.habilidade_canto,cao.nome)
```
