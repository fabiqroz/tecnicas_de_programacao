## Programação Orientada a Objetos (POO)

O que conhecemos por programação até agora se encaixa em um paradigma específico, que chamamos de **Programação Estruturada**

Programação estruturada (criada no final de 1950) é uma forma de programação que preconiza que todos os programas possíveis podem ser reduzidos a apenas três estruturas: sequência, decisão e iteração (repetição).

Tendo, na prática, sido transformada na programação modular, a programação estruturada orienta os programadores para a criação de estruturas simples nos programas,
usando funções.

Esta foi a forma dominante na criação de software anterior à programação orientada por objetos. 

Muitas linguagens de programação são estritamente estruturadas. Ex: *C, ascal, PHP, Cobol*, etc.

Nós podemos programar em Python seguindo um paradigma estruturado, e foi isso o que aprendemos na disciplinas de *Introdução a Computação*.

```python
# Funções (Procedimentos) criados por nós

def func_1(a, b):
    # Variáveis Locais
    x = a+b
    return x

def func_2(r):
    # Variáveis Locais
    a = PI*r**2
    return a

# Variáveis Globais
PI = 3.14
a = 1.0
b = 2

# Chamadas a funções
soma = func_1(a,b)
area = func_2(2)

# Funções Nativas
print(soma,area)
```

Contudo, a partir de agora, trazemos uma nova e importante informação: 
Apesar de ser possível se programar de forma estruturada,  **python é uma linguagem de programação orientada a objetos**.

Para entender melhor isto, vamos observar  algumas questões (que foram ignoradas em IC) com um pouco de atenção:

```python
print(type(1))
print(type(1.5))
print(type('abc'))
print(type(True))
print(type([1,2,3]))
```

**Tudo** no Python é definido utilizando-se o conceito  de classes e objetos.

Quando eu defino `a = 'abc'`, na verdade eu estou criando um **objeto** (que chamamos de `a`) que é uma representação de um dado da **classe** `str`(*string*).

Acima, nós afirmamos que objetos de uma classe possui características e comportamentos compartilhados. Podemos então pensar que objetos da classe `str` irão, portanto, ter características e comportamentos compartilhados entre si. 


### Comportamento de um objeto: Métodos

```python
a = 'Ola Mundo'
palavras_a = [a.split()]
print(palavras_a)
```

```python
b = 'Maria Jose'
palavras_b = [b.split()]
print(palavras_b)
```

```python
c = 17
numeros_c = [a.split()]
print(numeros_c)
```

Podemos observar que os objetos `a` e `b` são ambos instâncias da classe `str`, portanto ambos compartilham a função `split()`. 
Quando tentamos acessar essa mesma função em um objeto `c` da classe `int` isso não foi possível. 

Os comportamentos definem ações e estas ações (como nós já aprendemos) são implementadas a través de funções.
Quando um omportamento está associado APENAS a objetos de uma determinada classe, estas funções passam a ser chamadas de **métodos**. Dessa forma, `split()` é um método da classe `str`. 

### Características de um objeto: Atributos ou Variáveis

O método `split()` executa uma ação relativa ao comportamento de um objeto da classe `str`. Mas muito objetos de classes específicas podem também ter atributos (ou variáveis) exclusivos que irão representar características do objeto. 

```python
n_int = 3
print(n_int.numerator, n_int.denominator)
```

```python
s = '3'
print(s.numerator, s.denominator)
```

Podemos ter objetos de classes diferentes que possuem um mesmo atributo:

```python
print(n_int.__class__)
print(s.__class__)
print(1.0.__class__)
print(False.__class__)
print([].__class__)
```

### Aperfeiçoando a classe *Cachorro*: Adcionando novos atributos (não fixos)

<img src="Figures/objetos_classe_cachorro.png" alt= “” width=700 height=350>

```python
class cachorro:  
  # Construtor da classe cachorro
  def __init__(self, nome, raca, vacina, idade, peso, tutor):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    self.classe = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

### Criando novos objetos da classe *Cachorro*

```python
cao_2 = cachorro(nome ='Teobaldo', raca='Pintcher', vacina = False, idade= 5, peso=3.0, tutor = 'Jose')
cao_3 = cachorro('Caramelo', 'SRD', True, 10, 12.0, 'Maria')
```

```python
# Imprimindo o atributo "idade" dos objetos criados
print(cao_2.idade)
print(cao_3.idade)
```

```python
# A espécie é a mesma para ambos 
print(cao_2.classe)
print(cao_3.classe)
```

```python
print('A raça do '+ cao_2.nome+' é '+ cao_2.raca)
print('A raça do '+ cao_3.nome+' é '+ cao_3.raca)
```

```python
caes = [cao_2,cao_3]

for cao in caes:
  if not cao.status_vacina:
    print('Atenção {}. Seu cãozinho, o {}, precisa tomar vacina.'.format(cao.nome_tutor, cao.nome), end='\n')
  else:
    print('Parabéns {}! Seu cãozinho, o {}, está vacinado!'.format(cao.nome_tutor, cao.nome), end='\n')
```

Eu posso também criar valores padrão para os atributos de um objeto para o caso de algum atributo não ser passado ao construtor:

```python
class cachorro:  
  # Construtor da classe cachorro inicializando atributos automaticamente em sua lista de parametros
  def __init__(self, nome='Pituca', raca = 'Poodle', vacina= True, idade=12, peso=5.5, tutor='Fulaninho'):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

Dessa forma, posso criar um objeto sem especificar nenhum atributo:


```python
cao_4 = cachorro()
print(cao_4.nome, cao_4.raca, cao_4.status_vacina, cao_4.idade, cao_4.peso, cao_4.nome_tutor)
```

### Definindo os métodos da classe *Cachorro*

```python
class cachorro:  
  
  # Construtor da classe cachorro
  def __init__(self, nome='Pituca', raca = 'Poodle', vacina= True, idade=12, peso=5.5, tutor='Fulaninho'):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
  
  # Método latir da classe cachorro
  def latir(self):
    print('{} diz AU AU AU!'.format(self.nome))
```

```python
cao_6 = cachorro('Teobaldo', 'Pintcher', False, 5, 3.0, 'Jose')
cao_6.latir()
```

### Explorando a classe `str`

```python
# Criei um objeto da classe str (string)
frase = str('o rato roeu a roupa do rei de roma')
# Isso equivale a fazer  a = 'o rato roeu a roupa do rei de roma'

# split é um método da classe str
palavras = frase.split()
print(palavras)

# upper é métpdo da classe str
print(frase.upper())

#Outros métodos
print(frase.replace('rato', 'gato'))

n_frase= ', '.join(['o rato', 'muito danado que era', 'roeu a roupa do rei de roma'])
print(n_frase)
```

### Explorando a classe `list`

```python
# Criei um objeto da classe list (Listas)
lista = list([10,20,30])
# Isso equivale a fazer  lista = [10,20,30]

# append é um método da classe list
lista.append(5)
print(lista)

# sort é um método da classe list
lista.sort()
print(lista)
```
