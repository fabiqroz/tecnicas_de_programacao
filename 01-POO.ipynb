## Programao Orientada a Objetos (POO)

O que conhecemos por programao at agora se encaixa em um paradigma especfico, que chamamos de **Programao Estruturada**

Programao estruturada (criada no final de 1950)  uma forma de programao que preconiza que todos os programas possveis podem ser reduzidos a apenas trs estruturas: sequncia, deciso e iterao (repetio).

Tendo, na prtica, sido transformada na programao modular, a programao estruturada orienta os programadores para a criao de estruturas simples nos programas,
usando funes.

Esta foi a forma dominante na criao de software anterior  programao orientada por objetos. 

Muitas linguagens de programao so estritamente estruturadas. Ex: *C, ascal, PHP, Cobol*, etc.

Ns podemos programar em Python seguindo um paradigma estruturado, e foi isso o que aprendemos na disciplinas de *Introduo a Computao*.

```python
# Funes (Procedimentos) criados por ns

def func_1(a, b):
    # Variveis Locais
    x = a+b
    return x

def func_2(r):
    # Variveis Locais
    a = PI*r**2
    return a

# Variveis Globais
PI = 3.14
a = 1.0
b = 2

# Chamadas a funes
soma = func_1(a,b)
area = func_2(2)

# Funes Nativas
print(soma,area)
```

Contudo, a partir de agora, trazemos uma nova e importante informao: 
Apesar de ser possvel se programar de forma estruturada,  **python  uma linguagem de programao orientada a objetos**.

Para entender melhor isto, vamos observar  algumas questes (que foram ignoradas em IC) com um pouco de ateno:

```python
print(type(1))
print(type(1.5))
print(type('abc'))
print(type(True))
print(type([1,2,3]))
```

Quando eu defino `a = 'abc'`, na verdade eu estou criando um **objeto** (que chamamos de `a`) que  uma representao de um dado da **classe** `str`(*string*).

Acima, ns afirmamos que objetos de uma classe possui caractersticas e comportamentos compartilhados. Podemos ento pensar que objetos da classe `str` iro, portanto, ter caractersticas e comportamentos compartilhados entre si. 


### Comportamento de um objeto: Mtodos

```python
a = 'Ola Mundo'
palavras_a = [a.split()]
print(palavras_a)
```

```python
b = 'Maria Jose'
palavras_b = [b.split()]
print(palavras_b)
```

```python
c = 17
numeros_c = [a.split()]
print(numeros_c)
```

Podemos observar que os objetos `a` e `b` so ambos instncias da classe `str`, portanto ambos compartilham a funo `split()`. 
Quando tentamos acessar essa mesma funo em um objeto `c` da classe `int` isso no foi possvel. 

Os comportamentos definem aes e estas aes (como ns j aprendemos) so implementadas a travs de funes.
Quando um omportamento est associado APENAS a objetos de uma determinada classe, estas funes passam a ser chamadas de **mtodos**. Dessa forma, `split()`  um mtodo da classe `str`. 

### Caractersticas de um objeto: Atributos ou Variveis

O mtodo `split()` executa uma ao relativa ao comportamento de um objeto da classe `str`. Mas muito objetos de classes especficas podem tambm ter atributos (ou variveis) exclusivos que iro representar caractersticas do objeto. 

```python
n_int = 3
print(n_int.numerator, n_int.denominator)
```

```python
s = '3'
print(s.numerator, s.denominator)
```

Podemos ter objetos de classes diferentes que possuem um mesmo atributo:

```python
print(n_int.__class__)
print(s.__class__)
print(1.0.__class__)
print(False.__class__)
print([].__class__)
```

### Aperfeioando a classe *Cachorro*: Adcionando novos atributos (no fixos)

<img src="Figures/objetos_classe_cachorro.png" alt=  width=700 height=350>

```python
class cachorro:  
  # Construtor da classe cachorro
  def __init__(self, nome, raca, vacina, idade, peso, tutor):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    self.classe = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

### Criando novos objetos da classe *Cachorro*

```python
cao_2 = cachorro(nome ='Teobaldo', raca='Pintcher', vacina = False, idade= 5, peso=3.0, tutor = 'Jose')
cao_3 = cachorro('Caramelo', 'SRD', True, 10, 12.0, 'Maria')
```

```python
# Imprimindo o atributo "idade" dos objetos criados
print(cao_2.idade)
print(cao_3.idade)
```

```python
# A espcie  a mesma para ambos 
print(cao_2.classe)
print(cao_3.classe)
```

```python
print('A raa do '+ cao_2.nome+'  '+ cao_2.raca)
print('A raa do '+ cao_3.nome+'  '+ cao_3.raca)
```

```python
caes = [cao_2,cao_3]

for cao in caes:
  if not cao.status_vacina:
    print('Ateno {}. Seu cozinho, o {}, precisa tomar vacina.'.format(cao.nome_tutor, cao.nome), end='\n')
  else:
    print('Parabns {}! Seu cozinho, o {}, est vacinado!'.format(cao.nome_tutor, cao.nome), end='\n')
```

### Explorando a classe `str`

```python
# Criei um objeto da classe str (string)
frase = str('o rato roeu a roupa do rei de roma')
# Isso equivale a fazer  a = 'o rato roeu a roupa do rei de roma'

# split  um mtodo da classe str
palavras = frase.split()
print(palavras)

# upper  mtpdo da classe str
print(frase.upper())

#Outros mtodos
print(frase.replace('rato', 'gato'))

n_frase= ', '.join(['o rato', 'muito danado que era', 'roeu a roupa do rei de roma'])
print(n_frase)
```
