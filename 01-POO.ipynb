# Atributos de uma classe

5.png

A sintaxe para criar um atributo :
    
     self.attribute = algo
    
Existe um mtodo especial chamado:

     __init__()

Este mtodo, tambm chamado de **construtor**  usado para inicializar os atributos de um objeto. Por exemplo:

```python
class Cachorro:
  # Construtor da classe cachorro
  def __init__(self):
    # Aributo fixo da classe cachorro
    self.especie = 'mamifero'
```

```python
class Cachorro:  
  # Construtor da classe cachorro
  def __init__(self, nome, raca, vacina, idade, peso, tutor):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

Eu posso tambm criar valores padro para os atributos de um objeto para o caso de algum atributo no ser passado ao construtor:

```python
class Cachorro:  
  # Construtor da classe cachorro inicializando atributos automaticamente em sua lista de parametros
  def __init__(self, nome='Pituca', raca = 'Poodle', vacina= True, idade=12, peso=5.5, tutor='Fulaninho'):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

```python
cao_1.latir()
cao_2.latir()
cao_3.latir()
```

```python
# Criei um objeto da classe str (string)
frase = str('o rato roeu a roupa do rei de roma')
# Isso equivale a fazer  a = 'o rato roeu a roupa do rei de roma'

# upper  mtpdo da classe str
print(frase.upper())

# split  um mtodo da classe str
palavras = frase.split()
print(palavras)
```

```python

num_r = frase.count('r')
print(num_r)
```

Aqui estou chamando uma funo criada por mim, em que passo minha frase como parametro para contar o numero de erres

```python
def count(frase, letra):
  cont = 0
  for c in frase:
    if c == 'r':
      cont +=1
  return cont

num_r_2 = count(frase, 'r')
print(num_r_2)
```
