# Classes

**Classe** é um conjunto de objetos distintos, porém com as mesmas **características** e **comportamentos**. A classe pode ser vista como uma abstração de entidades reais ou abstratas. Ex:


1.   Classe *Pessoa*
2.   Classe *Animal*
3.   Classe *Automóvel*
4.   Classe *Número inteiro*
5.   Classe *Número reais*
6.   E por aí vai...

# Objetos 

**Objeto** é uma instância ou modelo derivado de uma classe. Portanto objeto também é a representação de qualquer coisa, real ou abstrata. Ex:


1.   Objetos da Classe *Pessoa*:
  1.   pessoa -> João
  2.   pessoa -> Maria
  3.   pesoa -> José
2. Objetos da classe *Animal*:
  1. animal -> cachorro
  2. animal-> gato
3. Objetos da classe *Numero Inteiro*
  1. 1
  2. 2
  3. 68058

# Atributos de uma classe

5.png

A sintaxe para criar um atributo é:
    
     self.attribute = algo
    
Existe um método especial chamado:

     __init__()

Este método, também chamado de **construtor** é usado para inicializar os atributos de um objeto. Por exemplo:

```python
class Cachorro:
  # Construtor da classe cachorro
  def __init__(self):
    # Aributo fixo da classe cachorro
    self.especie = 'mamifero'
```

```python
class Cachorro:  
  # Construtor da classe cachorro
  def __init__(self, nome, raca, vacina, idade, peso, tutor):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

```python
cao_1 = Cachorro(nome ='Teobaldo', raca='Pintcher', vacina = False, idade= 5, peso=3.0, tutor = 'Jose')
cao_2 = Cachorro('Caramelo', 'SRD', True, 10, 12.0, 'Maria')
```

```python
# Imprimindo o atributo "nome" dos objetos criados
print(cao_1.nome)
print(cao_2.nome)
```

```python
# Imprimindo o atributo "idade" dos objetos criados
print(cao_1.idade)
print(cao_2.idade)
```

```python
caes = [cao_1,cao_2]

for cao in caes:
  if not cao.status_vacina:
    print('Atenção {}. Seu cãozinho, o {}, precisa tomar vacina. Dê vacilo não!'.format(cao.nome_tutor, cao.nome), end='\n')
  else:
    print('Parabéns {}! Seu cãozinho, o {}, está vacinado!'.format(cao.nome_tutor, cao.nome), end='\n')
```

Eu posso também criar valores padrão para os atributos de um objeto para o caso de algum atributo não ser passado ao construtor:

```python
class Cachorro:  
  # Construtor da classe cachorro inicializando atributos automaticamente em sua lista de parametros
  def __init__(self, nome='Pituca', raca = 'Poodle', vacina= True, idade=12, peso=5.5, tutor='Fulaninho'):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
```

Dessa forma, posso criar um objeto sem especificar nenhum atributo:

```python
class Cachorro:  
  # Construtor da classe cachorro
  def __init__(self, nome='Pituca', raca = 'Poodle', vacina= True, idade=12, peso=5.5, tutor='Fulaninho'):
    # Atributo fixo da classe cachorro
    self.especie = 'mamifero'
    # Outros atributos da classe cachorro
    self.nome = nome
    self.raca = raca
    self.status_vacina = vacina
    self.idade = idade
    self.peso = peso
    self.nome_tutor = tutor
  
  # Método latir da classe cachorro
  def latir(self):
    print('{} diz AU AU!'.format(self.nome))
```

```python
cao_1 = Cachorro('Teobaldo', 'Pintcher', False, 5, 3.0, 'Jose')
cao_2 = Cachorro('Caramelo', 'SRD', True, 10, 12.0, 'Maria')
cao_3 = Cachorro()
```

```python
cao_1.latir()
cao_2.latir()
cao_3.latir()
```

```python
# Criei um objeto da classe str (string)
frase = str('o rato roeu a roupa do rei de roma')
# Isso equivale a fazer  a = 'o rato roeu a roupa do rei de roma'

# upper é métpdo da classe str
print(frase.upper())

# split é um método da classe str
palavras = frase.split()
print(palavras)
```

```python
# Criei um objeto da classe list (Listas)
lista = list([10,20,30])
# Isso equivale a fazer  lista = [10,20,30]

# append é um método da classe list
lista.append(5)
print(lista)

# sort é um método da classe list
lista.sort()
print(lista)
```

# Diferença entre métodos e funções

Aqui estou chamando um método da classe string que conta a quantodade de erres

```python

num_r = frase.count('r')
print(num_r)
```

Aqui estou chamando uma função criada por mim, em que passo minha frase como parametro para contar o numero de erres

```python
def count(frase, letra):
  cont = 0
  for c in frase:
    if c == 'r':
      cont +=1
  return cont

num_r_2 = count(frase, 'r')
print(num_r_2)
```
